```python
# Importo "Datos_perira" y lo converto en un nuevo DATAFRAME
Datos_pereira = pd.read_excel('/content/drive/MyDrive/Proyecto_personal/TerriData66001.xlsx')
Datos_pereira.head()
```

```python
# veo la informacion del Dataframe
Datos_pereira.info()
```

```python
# Elimino la primera fila que no tiene informacion
Datos_pereira.drop(0)
```

```python
# Elimino los valores nulos de la columna dato numerico
Datos_pereira_no_nulos = Datos_pereira.dropna(subset=['Dato Numérico'], how='any')
Datos_pereira_no_nulos.info()
```

```python
# cambio el nombre de la variable
dpn = Datos_pereira_no_nulos
```

```python
#convierto los valores de la columna Dato Numérico string a float
dpn['Dato Numérico'] = dpn['Dato Numérico'].str.replace('[\,\.]','')
dpn['Dato Numérico'] = dpn['Dato Numérico'].astype(float)/100
dpn.head()
```

```python
# observo los valores unicos de la columna unidad de medida
print(dpn['Unidad de Medida'].unique())
```

```python
# divido los valores de porcentaje en 100
dpn.loc[dpn['Unidad de Medida'] == 'Porcentaje', 'Dato Numérico'] = dpn.loc[dpn['Unidad de Medida'] == 'Porcentaje', 'Dato Numérico'] / 100
```

```python
# corroboro que todo este bien
filas_porcentaje = dpn.loc[dpn['Unidad de Medida'] == 'Porcentaje']

print(filas_porcentaje)
```

```python
# Se hace lo mismo para Porcentaje (el valor está multiplicado por 100)
dpn.loc[dpn['Unidad de Medida'] == 'Porcentaje (el valor está multiplicado por 100)', 'Dato Numérico'] = dpn.loc[dpn['Unidad de Medida'] == 'Porcentaje (el valor está multiplicado por 100)', 'Dato Numérico'] / 100

filas_porcentaje_x100 = dpn.loc[dpn['Unidad de Medida'] == 'Porcentaje (el valor está multiplicado por 100)']

print(filas_porcentaje_x100)
```

```python
# Elimino las comulumnas innecesarias
columnas_a_eliminar = ['Código Departamento', 'Departamento', 'Código Entidad']
dpn = dpn.drop(columnas_a_eliminar, axis=1)

dpn
```

```python
columnas = ['Dimensión', 'Subcategoría', 'Indicador', 'Dato Cualitativo', 'Fuente', 'Unidad de Medida']  # Lista de nombres de las columnas

# Elimino los espacios al inicio y al final de cada string							
for col in columnas:
    dpn[col] = dpn[col].str.strip()
```

```python
ruta_guardado = '/content/drive/MyDrive/Proyecto_personal/Datos_pereira.xlsx'
dpn.to_excel(ruta_guardado, index=False)
```
